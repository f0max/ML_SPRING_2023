```python
#df = pd.read_csv('Shoe prices.csv')
df = pd.read_csv('test.csv')
```

```python
total_rows = df.shape[0]
total_cols = df.shape[1]
print(f'Всего строк: {total_rows}')
print(f'Всего столбцов: {total_cols}')
```

```python
df_new = df.dropna(axis=1, how='any')
print(f'до: {df.shape}')
print(f'после: {df_new.shape}')
```

```python
df_new.isna().sum()
```

```python
num_cols = []
for col in df.columns:
    temp_null_count = df[df[col].isnull()].shape[0]
    dt = str(df[col].dtype)
    if temp_null_count>0 and (dt=='float64' or dt=='int64'):
        num_cols.append(col)
        temp_perc = round((temp_null_count / total_rows) * 100.0, 2)
        print(f'Колонка {col}. Тип данных {dt}. Количество пустых значений {temp_null_count}, {temp_perc}%.')
```

```python
columns_to_replace = ['LotFrontage', 'GarageYrBlt']
for column in columns_to_replace:
    df[column] = df[column].fillna(0)
```

```python
df.isna().sum()
```

```python
cols_to_drop = ['MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 'BsmtFullBath', 'GarageCars', 'GarageArea', 'MSZoning', 'SaleType']
df.dropna(subset=cols_to_drop,inplace = True, axis = 0)
df.isna().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
new_dummies = pd.get_dummies(df['saletype'], prefix='saletype')
new_data = pd.concat([df, new_dummies], axis=1)
new_data.drop('saletype', axis=1, inplace=True)
```
